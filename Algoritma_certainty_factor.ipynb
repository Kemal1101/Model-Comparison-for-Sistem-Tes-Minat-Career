```python
import pandas as pd
```

```python
# =========================================
# 1. IMPORT LIBRARY
# =========================================

file_survey = "Preprocessed_RIASEC_Dataset.csv"

df_survey = pd.read_csv(file_survey)


print("=== DATA SURVEY ===")
print(df_survey.head())
```

```python
# =========================================
# 3. FUNGSI NORMALISASI (LIKERT 1-5 → CF 0-1)
# =========================================
def likert_to_cf(x):
    return (x - 1) / 4  # 1 → 0, 5 → 1
```

```python
# =========================================
# 4. RULE (BOBOT PAKAR)
# =========================================
rules = {
    "G1": 0.8,  # suka analisis data
    "G2": 0.7,  # problem solving
    "G3": 0.9   # penelitian
}
```

```python
# =========================================
# 5. AMBIL DATA GEJALA (CONTOH KOLOM)
# Sesuaikan dengan dataset kamu
# =========================================
# Misalnya:
# I1 = analisis data
# I2 = problem solving
# I3 = penelitian

df_survey["G1_cf"] = df_survey["I1"].apply(likert_to_cf)
df_survey["G2_cf"] = df_survey["I2"].apply(likert_to_cf)
df_survey["G3_cf"] = df_survey["I3"].apply(likert_to_cf)
```

```python
# =========================================
# 6. HITUNG CF PER RULE
# =========================================
df_survey["CF1"] = df_survey["G1_cf"] * rules["G1"]
df_survey["CF2"] = df_survey["G2_cf"] * rules["G2"]
df_survey["CF3"] = df_survey["G3_cf"] * rules["G3"]
```

```python
# =========================================
# 7. FUNGSI KOMBINASI CF
# =========================================
def combine_cf(cf1, cf2):
    return cf1 + cf2 * (1 - cf1)
```

```python
# =========================================
# 8. HITUNG CF FINAL
# =========================================
cf_temp = combine_cf(df_survey["CF1"], df_survey["CF2"])
df_survey["CF_Investigative"] = combine_cf(cf_temp, df_survey["CF3"])
```

```python
# =========================================
# 9. INTERPRETASI HASIL
# =========================================
def interpret_cf(cf):
    if cf >= 0.8:
        return "Sangat Yakin"
    elif cf >= 0.6:
        return "Yakin"
    elif cf >= 0.4:
        return "Cukup"
    else:
        return "Rendah"

df_survey["Interpretasi"] = df_survey["CF_Investigative"].apply(interpret_cf)
```

```python
# =========================================
# 10. TAMPILKAN HASIL
# =========================================
print(df_survey[[
    "Nama Lengkap",
    "CF1", "CF2", "CF3",
    "CF_Investigative",
    "Interpretasi"
]])
```

```python
file_questions = "ONET_Interest_Profiler_Questions.xlsx"
file_jobs = "Onet_Top3_Occupations_by_InterestCode.xlsx"

df_questions = pd.read_excel(file_questions)
df_jobs = pd.read_excel(file_jobs)

print("=== DATA QUESTIONS ===")
print(df_questions.head())

print("\n=== DATA JOBS ===")
print(df_jobs.head())
```

```python
# =========================================
# HITUNG SKOR RIASEC DARI DATA SURVEY
# =========================================
df_survey["Skor_R"] = df_survey[["R1","R2","R3","R4","R5"]].mean(axis=1)
df_survey["Skor_I"] = df_survey[["I1","I2","I3","I4","I5"]].mean(axis=1)
df_survey["Skor_A"] = df_survey[["A1","A2","A3","A4","A5"]].mean(axis=1)
df_survey["Skor_S"] = df_survey[["S1","S2","S3","S4","S5"]].mean(axis=1)
df_survey["Skor_E"] = df_survey[["E1","E2","E3","E4","E5"]].mean(axis=1)
df_survey["Skor_C"] = df_survey[["C1","C2","C3","C4","C5"]].mean(axis=1)
```

```python
# =========================================
# GENERATE KODE RIASEC (TANPA DUMMY)
# =========================================
cols = ["Skor_R","Skor_I","Skor_A","Skor_S","Skor_E","Skor_C"]

def get_kode(row):
    # urutkan berdasarkan skor tertinggi
    sorted_cols = sorted(cols, key=lambda x: row[x], reverse=True)

    # ambil 3 tertinggi
    top3 = sorted_cols[:3]

    # ambil hurufnya saja (R, I, A, S, E, C)
    kode = "".join([col.split("_")[1] for col in top3])

    return kode

df_survey["Kode_RIASEC"] = df_survey.apply(get_kode, axis=1)
```

```python
# Merge berdasarkan kode RIASEC
df_result = df_survey.merge(
    df_jobs,
    left_on="Kode_RIASEC",
    right_on="Interest Code",
    how="left"
)
```

```python
print(df_result[[
    "Nama Lengkap",
    "CF_Investigative",
    "Kode_RIASEC",
    "Occupation"
]])
```

```python
df_survey["Kode_RIASEC"] = df_survey["Kode_RIASEC"].apply(lambda x: "-".join(list(x)))
```

```python
# =========================================
# 1. NORMALISASI DATA
# =========================================
df_result["Kode_RIASEC"] = df_result["Kode_RIASEC"].str.replace("-", "").str.upper()
df_jobs["Interest Code"] = df_jobs["Interest Code"].astype(str).str.upper()

# =========================================
# 2. FUNGSI SIMILARITY
# =========================================
def similarity(code1, code2):
    return len(set(code1) & set(code2))

# =========================================
# 3. FUNGSI CARI OCCUPATION (UPGRADE)
# =========================================
def cari_occupation(kode):
    kode = kode.upper()

    # hitung skor kecocokan
    df_jobs["score"] = df_jobs["Interest Code"].apply(
        lambda x: similarity(kode, str(x))
    )

    # urutkan berdasarkan skor tertinggi
    hasil = df_jobs.sort_values(by="score", ascending=False)

    # ambil 3 occupation terbaik (biar tidak monoton)
    return ", ".join(hasil["Occupation"].head(3))

# =========================================
# 4. ISI KOLOM OCCUPATION
# =========================================
df_result["Occupation"] = df_result["Kode_RIASEC"].apply(cari_occupation)

# =========================================
# 5. TAMPILKAN HASIL
# =========================================
print(df_result[[
    "Nama Lengkap",
    "CF_Investigative",
    "Kode_RIASEC",
    "Occupation"
]])
```

```python
# =========================================
# 1. NORMALISASI DATA
# =========================================
df_result["Kode_RIASEC"] = df_result["Kode_RIASEC"].str.replace("-", "").str.upper()
df_jobs["Interest Code"] = df_jobs["Interest Code"].astype(str).str.upper()

# =========================================
# 2. FUNGSI SIMILARITY
# =========================================
def similarity(code1, code2):
    return len(set(code1) & set(code2))

# =========================================
# 3. FUNGSI TOP 3 OCCUPATION — NAMA UNIK, TIDAK BERULANG
# =========================================
def cari_top3_occupation(kode):
    kode = kode.upper()

    df_temp = df_jobs.copy()
    df_temp["score"] = df_temp["Interest Code"].apply(
        lambda x: similarity(kode, str(x))
    )

    # Sort by score, lalu drop duplikat nama occupation
    hasil = (
        df_temp
        .sort_values(by="score", ascending=False)
        .drop_duplicates(subset="Occupation")   # <-- KUNCI: buang nama yang sama
    )

    top3 = hasil["Occupation"].head(3).tolist()

    # Padding kalau kurang dari 3
    while len(top3) < 3:
        top3.append("-")

    return pd.Series({
        "Occ_1": top3[0],
        "Occ_2": top3[1],
        "Occ_3": top3[2]
    })

# =========================================
# 4. APPLY KE DATAFRAME
# =========================================
df_result[["Occ_1", "Occ_2", "Occ_3"]] = df_result["Kode_RIASEC"].apply(cari_top3_occupation)

# =========================================
# 5. HILANGKAN DUPLIKAT BARIS (NAMA PESERTA)
# =========================================
df_result = df_result.drop_duplicates(subset="Nama Lengkap").reset_index(drop=True)

df_top3 = df_result[[
    "Nama Lengkap",
    "CF_Investigative",
    "Kode_RIASEC",
    "Occ_1",
    "Occ_2",
    "Occ_3"
]].head(3)

print(df_top3)

# Reset dulu sebelum apply
df_result = df_result.reset_index(drop=True)

# Baru apply
df_result[["Occ_1", "Occ_2", "Occ_3"]] = df_result["Kode_RIASEC"].apply(cari_top3_occupation)
```
